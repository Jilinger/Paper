# Einlesen

```python
global number_atoms
global radius
global r_b
coupling_constant= -2.72*10**9  #stärke der nachbarwechselwirkung
r_b = 2.5

#Parameter:
number_atoms=1000
densities= np.arange(0.01,0.53,0.01)

radiusse = (np.sqrt(number_atoms*r_b**2/densities))

iteration=5000
eigenvalues = np.empty((len(densities),number_atoms*iteration))
ratios_tmp =  np.empty((len(densities),number_atoms*iteration))
ipr =  np.empty((len(densities),number_atoms*iteration))
zeiten= ([[] for i in range(len(densities))])
gebrauchte_zeit = np.empty((len(densities)))
ratios =  np.empty((len(densities),(number_atoms-2)*iteration))

eigenvalues_ratios =  np.empty((len(densities),(number_atoms-2)*iteration))
for i,density in enumerate(densities):
    data = open("Results/Eigenstates/nr_atoms_"+str(number_atoms)+"/dens_"+str(np.round(density,3))+".npy","rb")
    #data = open("Results/Eigenstates/nr_atoms_"+str(number_atoms)+"/einmal_average/dens_"+str(np.round(density,3))+".npy","rb")
    text           = np.load(data)
    gebrauchte_zeit[i] = np.load(data)
    #zeiten[i]      = str(text)[95:-10]
    ipr[i]         = np.load(data)
    ratios_tmp[i]      = np.load(data)
    eigenvalues[i] = np.load(data)
    
    index= np.argwhere(ratios_tmp[i]==-1)
    ratios[i]=np.delete(ratios_tmp[i],index)
    eigenvalues_ratios[i]=np.delete(eigenvalues[i],index)
    
    
dens=np.zeros((len(densities)*number_atoms*iteration))
k=0
for density in densities:
    for i in range(number_atoms*iteration):
        dens[i+k*number_atoms*iteration]=density
    k+=1
    
b=np.zeros(number_atoms*iteration)
c,d=np.meshgrid(b,densities)
dens_mesh=d.ravel() 

for i in range(len(zeiten)):
        zeiten[i]=zeiten[i].replace("t","")
        zeiten[i]=zeiten[i].replace("=","")
zeiten = np.array(zeiten).astype(np.float)
```
